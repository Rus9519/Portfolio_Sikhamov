## Предобработка данных

### Работа с пропусками

Посмотрим на количество пропущенных значений в каждом столбце.

```python
df.isna().sum()
```

В столбцах *floors_total* и *locality_name* количество пропущенных значений ничтожно мало. Данные строки можно выбросить из набора данных, это не повлияет на результаты.

```python
df = df.dropna(subset=['floors_total', 'locality_name'])
```

```python
print('Удалено', 23699 - len(df), 'строки.')
```

Для обработки пропущенных значений в оставшихся столбцах, пройдемся по ним по порядку. 

#### Высота потолков

Начнем с высоты потолков. Посмотрим на распределение значений. В предыдущем параграфе отмечалось, что обнаружены потенциальные выбросы (высота 100 м). Для корректного отображения зададим диапазон до 10 метров.

```python
df.plot(y='ceiling_height', kind='hist', bins=50, range=(0,10))
plt.xlabel('ceiling_height, m')
plt.title('Ceiling height distribution')
plt.show()
```

Виден ярко выраженный пик на отметке около 2.5 метра, что логично, так как обычно квартиры имеют потолки высотой до 3х метров. В таком случае разумно заполнить данные медианным значением. Если пользователь не указал высоту потолков, можно предположить, что она примерно такая же, как и в остальных квартирах. Однако количество пропущенных значений чрезвычайно велико (около 40%), и заполнение их медианой сильно исказит выборку. В таком случае мы вынуждены оставить пропуски.

#### Жилая площадь

Следующим по списку идет столбец с жилой площадью. Для него стратегия заполнения медианой не подходит, так как неуказанное значение может быть абсолютно любым. 
Скорее всего, данные значения пропущены из-за того, что пользователь не увидел или не понял разницу между общей площадью и жилой площадью.
Предположим, что существует сильная зависимость жилой площади от общей площади.

```python
sns.scatterplot(data=df, x='total_area', y='living_area')
plt.text(x=600, y=50, s='Корреляция: {:.2f}'.format(df['living_area'].corr(df['total_area'])))
plt.title('Living area depending on total area')
plt.xlabel('total area, m^2')
plt.ylabel('living area, m^2')
plt.show()
```

Существует сильная корреляция между жилой и общей площадью.
Одним из возможных способов заполнения пропусков может быть создание линейной регрессии по имеющимся значениям и заполнение пропусков предсказанными моделью значениями. Однако, для упрощения, посмотрим, какую часть от общей составляет жилая площадь. Этим значением и заполним пропуски. Стоит отметить, что в столбце "общая площадь" нет пропусков, что делает оба описанных выше метода применимыми.

```python
(df.dropna()['living_area'] / df.dropna()['total_area']).mean()
```

```python
df.loc[df['living_area'].isna(), 'living_area'] = df['total_area'] * 0.57
```

#### Апартаменты

В столбце *is_apartment* пропущено огромное количество значений. Посмотрим на то, что в столбце все же присутствует.

```python
df.is_apartment.value_counts().to_frame()
```

Видим, что большинство объявлений не являются апартаментами.
Мы знаем, что апартаменты — это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания. В данной работе мы исследуем данные о продаже квартир, а квартира по определению является жилым помещением (как правило). В данном случае все пропущенные значения можно смело заменить на "False".

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

#### Площадь кухни

Как и в случае с жилой площадью, площадь кухни может зависеть от общей площади. Посмотрим на диаграмму рассеяния и корреляцию.

```python
sns.scatterplot(data=df, x='total_area', y='kitchen_area')
plt.text(x=600, y=45, s='Корреляция: {:.2f}'.format(df['kitchen_area'].corr(df['total_area'])))
plt.title('Kitchen area depending on total area')
plt.xlabel('total area, m^2')
plt.ylabel('kitchen area, m^2')
plt.show()
```

Видно, что имеется зависимость, однако корреляция средняя, здесь всё не так однозначно, как в случае с жилой площадью.

Заполнять данные пропуски медианой по всему датафрейму здесь неправильно, так как всё же присутствует корреляция с жилой площадью. В таком случае можно воспользоваться следующей стратегией: разделим все квартиры на 4 группы по величине общей площади и заполним пропуски медианным значением для каждой группы. Для начала посмотрим на распределения кухонной площади в каждой группе.

```python
plt.figure(figsize=(10, 3.5))
sns.kdeplot(df.dropna().query('total_area < 50')['kitchen_area'], label='total_area < 50', color='blue')
sns.kdeplot(df.dropna().query('50 <= total_area < 100')['kitchen_area'], label='50 <= total_area < 100', color='red')
plt.ylim(0, 0.2)
plt.xlim(0, 50)
plt.title('Kitchen are distribution for different groups of total area')
plt.xlabel('kitchen_area, m^2')
plt.legend()

plt.figure(figsize=(10, 3.5))
sns.kdeplot(df.dropna().query('100 <= total_area < 150')['kitchen_area'], label='100 <= total_area < 150', color='green')
sns.kdeplot(df.dropna().query('150 <= total_area')['kitchen_area'], label='total_area > 150', color='black')
plt.ylim(0, 0.2)
plt.xlim(0, 50)
plt.xlabel('kitchen_area, m^2')
plt.legend()
plt.show()
```

Отчетливо видно, что площадь кухни растет с увеличением общей площади. Для каждой следующей группы пик становится все менее выраженным, это связано с тем, что в группах находится разное количество квартир.

```python
median_1 = df.dropna().query('total_area < 50')['kitchen_area'].median()
median_2 = df.dropna().query('50 <= total_area < 100')['kitchen_area'].median()
median_3 = df.dropna().query('100 <= total_area < 150')['kitchen_area'].median()
median_4 = df.dropna().query('total_area >= 150')['kitchen_area'].median()
```

```python
df.loc[(df['kitchen_area'].isna()) & (df['total_area'] < 50), 'kitchen_area'] = median_1
df.loc[(df['kitchen_area'].isna()) & (50 <= df['total_area']) & (df['total_area'] < 100), 'kitchen_area'] = median_2
df.loc[(df['kitchen_area'].isna()) & (100 <= df['total_area']) & (df['total_area'] < 150), 'kitchen_area'] = median_3
df.loc[(df['kitchen_area'].isna()) & (df['total_area'] >= 150), 'kitchen_area'] = median_4
```

#### Балкон

В столбце *balcony* пропущено большо количество значений. Предполагаем, что если количество балконов не указано, значит оно равно нулю.

```python
df['balcony'] = df['balcony'].fillna(0)
```

#### Ближайшие аэропорт, центр города, парк и водоём

Скорее всего, данные получены автоматически на основе картографических данных. В таком случае причиной пропусков может быть:
* Ошибка в алгоритме записи данных,
* Данные объекты могут лежать на расстоянии, которое превышает установленную в системе зону поиска.

В данном случае сложно выбрать разумную систему заполнения пропусков. Если заполним нулями, это будет значить, что все эти объекты лежат вплотную к квартирам. Если заполним неким максимально возможным или медианным значением, то это не будет отражать реальную картину. В принципе можно удалить строки с пропущенными значениями, но тогда мы потеряем огромную часть данных. В таком случае, самым рациональным выбором будет оставить пропуски. 

#### Количество парков и водоёмов в радиусе 3 км

Количество пропусков в этих столбцах равно. Сначала проверим, нет ли взаимосвязи.

```python
len(df[(df['parks_around3000'].isna()) & (~df['ponds_around3000'].isna())])
```

Взаимосвязь обнаружена. Если не указано количество парков, то и не указано количество водоёмов в округе.

Скорее всего пропуски означают, что в радиусе 3 км нет ни того, ни другого. В таком случае заполняем пропуски нулями.

```python
df['parks_around3000'] = df['parks_around3000'].fillna(0)
df['ponds_around3000'] = df['parks_around3000'].fillna(0)
```

#### Продолжительность размещения объявления

Сложно сказать, что стало причиной появления пропусков в данном столбце. 

Также сложно выделить какие-либо взаимосвязи с другими данными. Заполнение медианой может сильно исказить данные. Построим 2 распределения, одно по имеющимся данным в данном столбце. Другое по данными, где пропуски заменим медианой.

```python
_, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15,5))
ax1.hist(df['days_exposition'], bins=50, range=(0,365))
ax2.hist(df['days_exposition'].fillna(df['days_exposition'].dropna().median()), bins=50, range=(0,365))
ax1.set_title('Days exposition distribution for 1 year')
ax2.set_title('Days exposition distribution for 1 year with the missing values filled by the median')
ax1.set_xlabel('days exposition')
ax2.set_xlabel('days exposition')
ax1.set_ylabel('frequency')
ax2.set_ylabel('frequency')
plt.show()
```

Данная величина имеет Пуассоновское распределение. Заполнение пропусков медианным значением сильно искажает исходные данные. В таком случае лучше оставить пропуски и ничего с ними не делать, так как удалив более 3000 строк мы лишимся большой доли данных, что может негативно сказаться на исследовательском анализе данных

#### Вывод по работе с пропусками

* 134 строки были удалены
* Пропуски в столбцах высота потолков, ближайший аэропорт, центр города, пруд и парк, а также в столбце продолжительности размещения объявления были оставлены без изменений в виду невозможности их корректного заполения. Удаление же строк приведет к потере большого количества данных

## Расчёты и добавление результатов в таблицу

```python
df['exposition_day'] = df['first_day_exposition'].dt.weekday
df['exposition_month'] = df['first_day_exposition'].dt.month
df['exposition_year'] = df['first_day_exposition'].dt.year
```

Добавим этаж квартиры. Группы - первый, последний, другой.

```python
def floor_transformer(row):
    '''
    Данная функция выполняет категоризацию по этажам квартир.
    '''
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'
```

Для удобства дальнейших сравнений напишем функцию. Параметр *dropna* будет нужен для столбцов, где мы намеренно оставили пропущенные значения.

```python
def plot_difference(column):
    '''
    Данная функция строит распределения по выбранному столбцу по данным в центре города и за его пределами.
    '''
    plt.figure(figsize=(12,4))
    sns.kdeplot(df_center.dropna()[column], 
                label='center')
    sns.kdeplot(df_cleaned.dropna().query('locality_name == "Санкт-Петербург" and distance_center_km > 5')[column],
                label='the rest of the city')
    plt.legend()
    plt.title(f'Distribution of {column} in the center and in the rest of the city')
    plt.show()
```

#### Факторы, влияющие на цену

##### Число комнат

Чтобы оценить влияние количества комнат на цену возьмем медианное значение для каждого количества комнат.

```python
sns.barplot(data=df_center, x='rooms', y='last_price', estimator=np.median, ci=None)
plt.title('Median price over the number of rooms')
plt.ylabel('median price')
plt.show()
```

Как и в случае со всем набором данных, цена растет по мере увеличения количества комнат, и с ростом их количества влияние ослабляется. Отличие от всего набора данных лишь в том, что в центре Санкт-Петербурга нет разницы между 5 и 6 комнатами, а по всему набору данных нет разница между 4 и 5 комнатами. 

Оценим медиану цены для разного количества комнат в центре города и в остальном городе.

```python
def plot_difference_bar(column):
    '''
    Данная функция принимает интересующий нас столбец.
    А выдает столбчатую диаграмму со сравнением медиан интересующего нас столбца в центре города и за его пределами.
    '''
    rest_city = (df_cleaned
                 .query('locality_name == "Санкт-Петербург" and distance_center_km > 5')
                 .groupby(column)['last_price'].median())

    center = df_center.groupby(column)['last_price'].median()

    labels = rest_city.index
    x = np.arange(len(labels))
    width = 0.35

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, rest_city, width, label='rest_city')
    rects2 = ax.bar(x + width/2, center, width, label='center')

    ax.set_ylabel('median_price')
    ax.set_xlabel(column)
    ax.set_title(f'Influence of {column} on the median price')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    plt.ylim(0, 1.2 * center.max() )

    plt.show()
```

```python
plot_difference_bar('rooms')
```

Не беря в учет минорные различия, тренд одинаковый: чем больше комнат, тем дороже квартира.
